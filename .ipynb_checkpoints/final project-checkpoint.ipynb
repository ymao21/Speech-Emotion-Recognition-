## 2. Load and organize data

The RAVDESS dataset stores each recording as a `.wav` file inside actor-specific folders (e.g., `Actor_01/`, `Actor_02/`, вЂ¦). The emotion label is **encoded in the filename** rather than provided in a separate CSV, which is unusual but common for audio datasets.

A typical filename looks like:

```
03-01-05-01-02-01-12.wav
```

Each of the seven hyphen-separated parts means something:

| Position | Meaning                | Example  |
|----------|------------------------|----------|
| 1        | Modality (03 = audio)  | 03       |
| 2        | Vocal channel          | 01 = speech |
| 3        | **Emotion**            | 05 = angry  |
| 4        | Emotional intensity    | 01 = normal |
| 5        | Statement              | 02       |
| 6        | Repetition             | 01       |
| 7        | Actor ID (odd = male, even = female) | 12 |

We only need the **emotion code** and the **actor ID**, but parsing all of it now makes the dataframe more flexible for downstream experiments.

Before we run this cell, we set `DATA_DIR` to wherever the dataset lives on disk. If you downloaded it from Kaggle, the extracted folder is usually named something like `audio_speech_actors_01-24` with subfolders `Actor_01 вЂ¦ Actor_24`.

````python
# === EDIT THIS PATH to point to your RAVDESS folder ===
# Examples:
#   DATA_DIR = "/Users/yourname/Downloads/audio_speech_actors_01-24"
#   DATA_DIR = "./data/ravdess"
DATA_DIR = "./data/ravdess"

# Mapping from RAVDESS emotion code -> human-readable label
EMOTION_MAP = {
    "01": "neutral",
    "02": "calm",
    "03": "happy",
    "04": "sad",
    "05": "angry",
    "06": "fearful",
    "07": "disgust",
    "08": "surprised",
}

def parse_ravdess_filename(filepath):
    """Pull the emotion code, emotion label, and actor ID out of a RAVDESS .wav filename.

    Returns a dict with keys: emotion_code, emotion_label, actor_id.
    Returns None if the filename doesn't follow the RAVDESS pattern.
    """
    name = Path(filepath).stem  # filename without .wav
    parts = name.split("-")
    if len(parts) != 7:
        return None
    emotion_code = parts[2]
    actor_id = parts[6]
    emotion_label = EMOTION_MAP.get(emotion_code)
    if emotion_label is None:
        return None
    return {
        "emotion_code": emotion_code,
        "emotion_label": emotion_label,
        "actor_id": actor_id,
    }

# Recursively find every .wav file under DATA_DIR
wav_paths = sorted(glob.glob(os.path.join(DATA_DIR, "**", "*.wav"), recursive=True))
print(f"Found {len(wav_paths)} .wav files in {DATA_DIR}")

# Build the dataframe
rows = []
for p in wav_paths:
    parsed = parse_ravdess_filename(p)
    if parsed is None:
        # We keep these around for the cleaning section so we can show how many
        # we filtered out and why.
        rows.append({
            "file_path": p,
            "emotion_code": None,
            "emotion_label": None,
            "actor_id": None,
        })
        continue
    rows.append({"file_path": p, **parsed})

df = pd.DataFrame(rows)
print(f"\nDataframe shape: {df.shape}")
df.head()
````

We print the dataframe shape and the first few rows to verify it looks right. We should see one row per `.wav` file with the parsed emotion and actor columns filled in.

````python
# Quick sanity look at the data we just built
print("Columns:", list(df.columns))
print("\nDtypes:\n", df.dtypes)
print("\nFirst 5 rows:")
df.head()
````

## 3. Data cleaning and validation

Before training a model, we want to make sure the dataset is actually clean and consistent. Even a few corrupted files or mis-named files can hurt training, and worse, they can produce silent bugs (the model "works" but on bad data). In this section we run a series of checks:

1. Are all files actually `.wav` files?
2. Do all filenames follow the expected RAVDESS 7-part structure?
3. Are there any invalid emotion codes (i.e., codes not in 01вЂ“08)?
4. Are there missing labels (rows where parsing failed)?
5. Are there duplicate file paths?
6. Can librosa actually load every file without errors?
7. Later, after feature extraction, we will also check for `NaN` or `inf` feature values.

For each check we either log a count, or remove the offending rows, and we always explain what we did.

````python
# --- Check 1: every file path ends with .wav ---
non_wav = df[~df["file_path"].str.lower().str.endswith(".wav")]
print(f"Files that are NOT .wav: {len(non_wav)}")
````

````python
# --- Check 2: filenames that don't follow the RAVDESS 7-part structure ---
# These are the rows where parse_ravdess_filename returned None (emotion_code is NaN).
bad_format = df[df["emotion_code"].isna()]
print(f"Filenames not following RAVDESS structure: {len(bad_format)}")
if len(bad_format) > 0:
    print("Examples:")
    print(bad_format["file_path"].head().tolist())
````

````python
# --- Check 3: invalid emotion codes ---
# Any code that exists in the data but is NOT in our known mapping.
valid_codes = set(EMOTION_MAP.keys())
invalid_codes = df.dropna(subset=["emotion_code"])
invalid_codes = invalid_codes[~invalid_codes["emotion_code"].isin(valid_codes)]
print(f"Rows with invalid emotion codes: {len(invalid_codes)}")
````

````python
# --- Check 4: missing labels ---
missing_label = df[df["emotion_label"].isna()]
print(f"Rows with missing emotion_label: {len(missing_label)}")
````

````python
# --- Check 5: duplicate file paths ---
dupes = df[df.duplicated(subset=["file_path"], keep=False)]
print(f"Duplicate file paths: {len(dupes)}")
````

We now drop any rows that failed the structural checks (rows 2 and 4). We keep a copy of the original dataframe in case we want to compare numbers later, but for the rest of the notebook we will work with the cleaned version.

````python
# Save the raw count, then drop unparseable rows
n_before = len(df)
df = df.dropna(subset=["emotion_code", "emotion_label", "actor_id"]).reset_index(drop=True)
n_after = len(df)
print(f"Dropped {n_before - n_after} unparseable rows. Clean rows: {n_after}")
````

````python
# --- Check 6: every file can be opened by librosa ---
# This is the slowest check, but it's the one that protects us most reliably.
# If a file is truncated/corrupted, librosa.load will raise an exception, and
# we drop it from the dataframe so it never sneaks into training.
bad_files = []
for p in df["file_path"]:
    try:
        # We don't need the audio yet, just confirm it loads. Using sr=None
        # preserves the file's native sampling rate, which we'll inspect later.
        y, sr = librosa.load(p, sr=None, duration=0.1)  # tiny slice = fast check
    except Exception as e:
        bad_files.append((p, str(e)))

print(f"Files that failed to load: {len(bad_files)}")
for p, err in bad_files[:5]:
    print(" -", p, "::", err)

if bad_files:
    bad_set = {p for p, _ in bad_files}
    df = df[~df["file_path"].isin(bad_set)].reset_index(drop=True)
    print(f"After removing unloadable files: {len(df)} rows remain.")
````

**Data cleaning summary.** After all checks, we keep only rows where (a) the filename follows the RAVDESS 7-part structure, (b) the emotion code is one of the 8 valid codes, and (c) librosa can actually open the file. We will run one final check on feature values (NaN / inf) right after feature extraction, since those only appear once we compute MFCCs.

### 4D. Audio duration analysis

Different emotions might be expressed at different speeds (e.g., excited speech tends to be faster). We want to know how long clips are on average so we can:

- decide whether trimming/padding is needed for sequence models later,
- catch unusually short or long clips that might be noise.

````python
# Compute the duration of every clip. This takes a few seconds on full RAVDESS.
durations = []
for p in df["file_path"]:
    try:
        # librosa.get_duration is much faster than loading the full waveform
        d = librosa.get_duration(path=p)
    except Exception:
        d = np.nan
    durations.append(d)

df["duration_sec"] = durations
print(df["duration_sec"].describe())
````

````python
plt.figure(figsize=(8, 4))
sns.histplot(df["duration_sec"].dropna(), bins=40, kde=True, color="steelblue")
plt.title("Distribution of clip durations (seconds)")
plt.xlabel("Duration (s)")
plt.ylabel("Number of clips")
plt.tight_layout()
plt.show()

print("Shortest clips:")
print(df.nsmallest(5, "duration_sec")[["file_path", "emotion_label", "duration_sec"]])
print("\nLongest clips:")
print(df.nlargest(5, "duration_sec")[["file_path", "emotion_label", "duration_sec"]])
````

**What we see.** RAVDESS clips are typically a few seconds long (most cluster around 3вЂ“4 seconds). For this baseline notebook we average MFCC features across time, so duration doesn't directly matter, but for the LSTM / CNN-LSTM teammates this is useful: they will likely want to pad or truncate to a fixed length.

### 4E. Sampling rate analysis

The sampling rate is how many audio samples per second the file contains (e.g., 48,000 Hz). If different clips have different sampling rates, MFCC features extracted from them won't be comparable, because the frequency content is sampled at different resolutions. We confirm the rate is consistent across the dataset.

````python
# Sample the first 100 files to keep this quick вЂ” we just want to confirm consistency
sample = df.sample(min(100, len(df)), random_state=SEED)
rates = []
for p in sample["file_path"]:
    try:
        y, sr = librosa.load(p, sr=None, duration=0.1)
        rates.append(sr)
    except Exception:
        pass

print("Unique sampling rates observed in a 100-file sample:", set(rates))
````

RAVDESS uses **48,000 Hz** throughout. When we extract MFCCs we will downsample to **22,050 Hz** (librosa's default) вЂ” this is plenty for speech and roughly halves the data we have to process. Consistency is what matters, not the exact rate.

### 4F. Waveform visualization

A waveform is just a plot of audio amplitude over time вЂ” it's the most "raw" view of a sound. Different emotions usually have different amplitude envelopes: angry speech often has bigger swings; sad speech tends to be quieter and more even. We plot one example per emotion to build intuition.

````python
def plot_waveform(filepath, title):
    y, sr = librosa.load(filepath, sr=None)
    plt.figure(figsize=(9, 2.5))
    librosa.display.waveshow(y, sr=sr, color="steelblue")
    plt.title(title)
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.tight_layout()
    plt.show()

# One example clip per emotion
example_clips = df.groupby("emotion_label").first().reset_index()
for _, row in example_clips.iterrows():
    plot_waveform(row["file_path"], f"Waveform вЂ” {row['emotion_label']}")
````

### 4G. Spectrogram visualization

A waveform shows amplitude over time, but it hides which **frequencies** are present. A spectrogram shows how energy is distributed across frequencies over time вЂ” it's basically a 2D heatmap where the x-axis is time, the y-axis is frequency, and color is intensity (in decibels). Speech recognition systems usually work on something like this rather than the raw waveform.

````python
def plot_spectrogram(filepath, title):
    y, sr = librosa.load(filepath, sr=None)
    # Short-time Fourier transform, then convert magnitude to decibels
    S = np.abs(librosa.stft(y))
    S_db = librosa.amplitude_to_db(S, ref=np.max)

    plt.figure(figsize=(9, 3))
    librosa.display.specshow(S_db, sr=sr, x_axis="time", y_axis="log", cmap="magma")
    plt.colorbar(format="%+2.0f dB")
    plt.title(title)
    plt.tight_layout()
    plt.show()

# Same example clips as above so we can compare to the waveforms
for _, row in example_clips.iterrows():
    plot_spectrogram(row["file_path"], f"Spectrogram вЂ” {row['emotion_label']}")
````

### 4H. MFCC visualization

**MFCCs** (Mel-frequency cepstral coefficients) are a compact summary of the spectrogram that emphasizes the frequency bands the human ear is most sensitive to. They're the most common audio feature for speech tasks because they capture the *shape* of the spectrum (which encodes things like vowel sounds and emotional tone) without all the noise of the full spectrogram.

Below, each column is a short time frame, each row is one of 40 MFCC coefficients, and the color shows the value. Different emotions tend to produce visibly different MFCC patterns.

````python
def plot_mfcc(filepath, title, n_mfcc=40):
    y, sr = librosa.load(filepath, sr=22050)
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
    plt.figure(figsize=(9, 3))
    librosa.display.specshow(mfccs, sr=sr, x_axis="time", cmap="coolwarm")
    plt.colorbar()
    plt.title(title)
    plt.ylabel("MFCC coefficient")
    plt.tight_layout()
    plt.show()

for _, row in example_clips.iterrows():
    plot_mfcc(row["file_path"], f"MFCC вЂ” {row['emotion_label']}")
````

### 4I. Feature distribution analysis

After we extract MFCCs (next section), it helps to look at how a few of the coefficients are distributed across emotions. If a coefficient's distribution looks very similar for every emotion, it probably won't help the model much. If it's clearly shifted between emotions, that's a useful signal.

We will revisit this after the feature extraction step.

### 4J. Correlation analysis

MFCC coefficients are not independent of each other вЂ” neighboring coefficients especially tend to be correlated. A correlation heatmap will tell us how much information is redundant. High correlation isn't a deal-breaker, but it means our 40-dimensional feature vector is "effectively" smaller than 40.

We will also revisit this after feature extraction.

## 5. Feature extraction

We now turn each audio clip into a fixed-length numerical vector that ML models can consume. The standard approach for a baseline is:

1. Load the audio with librosa.
2. Compute 40 MFCC coefficients across time (this gives a matrix of shape `(40, n_frames)`).
3. **Average the MFCCs across the time axis** to collapse the matrix into a single 40-dimensional vector.

> We average MFCCs across time because traditional machine learning models like Logistic Regression and Random Forests expect a *fixed-length* numerical input per sample. Audio clips have different lengths, so without this collapse step our feature vectors would have different shapes. Later, the LSTM/CNN-LSTM models will keep the time axis intact to exploit temporal information.

We wrap this in a function so we can call it cleanly for every file, and we wrap each call in a try/except so a single bad file doesn't crash the whole loop.

````python
N_MFCC = 40           # number of MFCC coefficients
SAMPLE_RATE = 22050   # resample everything to this rate for consistency

def extract_mfcc_mean(filepath, n_mfcc=N_MFCC, sr=SAMPLE_RATE):
    """Load an audio file and return the time-averaged MFCC feature vector.

    Returns a 1-D numpy array of length n_mfcc, or None if the file fails to load.
    """
    try:
        y, _ = librosa.load(filepath, sr=sr)
        # mfccs has shape (n_mfcc, n_frames). Mean across axis=1 collapses time.
        mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
        mfccs_mean = np.mean(mfccs, axis=1)
        return mfccs_mean
    except Exception as e:
        print(f"Failed to extract from {filepath}: {e}")
        return None

# Quick smoke test on one file to confirm the function works
test_vec = extract_mfcc_mean(df["file_path"].iloc[0])
print("Feature vector shape:", test_vec.shape if test_vec is not None else None)
print("First 5 values:", test_vec[:5] if test_vec is not None else None)
````

Now we run this for every file in the cleaned dataframe. This is the slowest cell in the notebook (a few minutes on full RAVDESS), so we add a small progress indicator so it doesn't feel frozen.

````python
features = []
labels = []
failed_paths = []

for i, row in df.iterrows():
    vec = extract_mfcc_mean(row["file_path"])
    if vec is None:
        failed_paths.append(row["file_path"])
        continue
    features.append(vec)
    labels.append(row["emotion_label"])
    if (i + 1) % 200 == 0:
        print(f"  processed {i + 1} / {len(df)} files")

X = np.array(features)
y = np.array(labels)
print(f"\nFinished. X shape: {X.shape}, y shape: {y.shape}")
print(f"Files that failed during feature extraction: {len(failed_paths)}")
````

````python
# Final cleaning step we promised earlier: NaN / inf in feature values.
nan_mask = np.isnan(X).any(axis=1) | np.isinf(X).any(axis=1)
print(f"Rows with NaN or inf in features: {nan_mask.sum()}")

if nan_mask.any():
    X = X[~nan_mask]
    y = y[~nan_mask]
    print(f"After removing them: X shape: {X.shape}, y shape: {y.shape}")
````

Now that we have a feature matrix, let's revisit the **feature distribution** and **correlation** analyses from EDA. We turn `X` into a dataframe so plotting is easier.

````python
mfcc_cols = [f"mfcc_{i}" for i in range(X.shape[1])]
feat_df = pd.DataFrame(X, columns=mfcc_cols)
feat_df["emotion_label"] = y
feat_df.head()
````

````python
# Distribution of a few selected MFCCs across emotions.
# We pick the first few because they tend to carry the most energy.
selected = ["mfcc_0", "mfcc_1", "mfcc_2", "mfcc_3"]
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
for ax, col in zip(axes, selected):
    sns.boxplot(data=feat_df, x="emotion_label", y=col, ax=ax, palette="Set3")
    ax.set_title(col)
    ax.tick_params(axis="x", rotation=45)
plt.tight_layout()
plt.show()
````

**What we see.** Some MFCC coefficients (especially the first few) show clearly different distributions across emotions вЂ” for example, the means tend to shift for angry vs. sad. That's the signal our models will pick up on.

````python
# Correlation heatmap among the 40 MFCC features.
plt.figure(figsize=(9, 7))
sns.heatmap(feat_df[mfcc_cols].corr(), cmap="coolwarm", center=0,
            xticklabels=False, yticklabels=False)
plt.title("Correlation between MFCC features")
plt.tight_layout()
plt.show()
````

**What we see.** There is moderate correlation between neighboring MFCCs (the bright diagonal stripe), which is expected: nearby coefficients describe similar parts of the cepstrum. There's no extreme redundancy though, so we will keep all 40 features.

## 6. Preprocessing

Three things happen before modeling:

1. **Encode the labels** from strings (`"angry"`, `"happy"`, вЂ¦) to integers вЂ” most ML models and Keras' `sparse_categorical_crossentropy` expect integer labels.
2. **Split** into a training set (80%) and a test set (20%), with **stratification** so each emotion is represented in both splits in the same proportions.
3. **Scale the features** so each MFCC has mean 0 and standard deviation 1 вЂ” this helps Logistic Regression and neural networks converge much faster.

A really important subtlety here is **data leakage**:

> We fit the scaler **only on the training data** so information from the test set does not accidentally leak into training. If we fit the scaler on the whole dataset before splitting, the means and standard deviations would already "know" about the test samples, and our test accuracy would be slightly optimistic.

There is also a second, more subtle form of leakage: **speaker leakage**. RAVDESS has only 24 actors, and a random split will almost certainly put the same speaker into both train and test. The model can then learn to recognize the *speaker's voice* rather than the *emotion*. We discuss this below and provide an actor-aware split as an option for teammates working on later models.

````python
# --- 1. Label encoding ---
le = LabelEncoder()
y_enc = le.fit_transform(y)

label_to_int = dict(zip(le.classes_, le.transform(le.classes_)))
int_to_label = {v: k for k, v in label_to_int.items()}
print("Label -> integer mapping:")
for lab, i in label_to_int.items():
    print(f"  {i}: {lab}")
````

````python
# --- 2. Stratified 80/20 split ---
X_train, X_test, y_train, y_test = train_test_split(
    X, y_enc,
    test_size=0.20,
    stratify=y_enc,    # keeps class proportions the same in train and test
    random_state=SEED,
)
print(f"Train: {X_train.shape}, Test: {X_test.shape}")

# Quick check that stratification worked
print("\nClass proportions in training set:")
print(pd.Series(y_train).value_counts(normalize=True).round(3).sort_index())
print("\nClass proportions in test set:")
print(pd.Series(y_test).value_counts(normalize=True).round(3).sort_index())
````

````python
# --- 3. Feature scaling (fit on train only!) ---
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)   # learns mean/std from training
X_test_scaled = scaler.transform(X_test)         # applies them to test
print("Scaling complete. Mean of training features:", X_train_scaled.mean().round(4),
      "Std:", X_train_scaled.std().round(4))
````

### Optional: actor-aware split

The split above is *random* over rows. If we want to test whether the model can generalize to *unseen speakers*, we should split by `actor_id` so that no actor appears in both train and test. The code below is provided as an option for downstream teammates вЂ” we don't use it in the baseline below to stay consistent with the project description, but it's a more honest evaluation of generalization.

````python
def actor_aware_split(df, X, y_enc, test_actor_frac=0.2, seed=SEED):
    """Split so that the same actor never appears in both train and test."""
    rng = np.random.default_rng(seed)
    actors = df["actor_id"].unique()
    rng.shuffle(actors)
    n_test = max(1, int(len(actors) * test_actor_frac))
    test_actors = set(actors[:n_test])
    train_mask = ~df["actor_id"].isin(test_actors).values
    test_mask = ~train_mask
    return (X[train_mask], X[test_mask], y_enc[train_mask], y_enc[test_mask],
            sorted(test_actors))

# Example (not used in the baseline below вЂ” uncomment if you want to try it):
# X_tr, X_te, y_tr, y_te, held_out_actors = actor_aware_split(df, X, y_enc)
# print("Held-out test actors:", held_out_actors)
````

## 7. Model 1 вЂ” traditional ML baseline

A **baseline** model is the simplest reasonable model we can throw at the problem. Its job isn't to be the best вЂ” its job is to give us a number to compare against. If our fancier deep learning model can't beat the baseline, something is wrong (with the features, the architecture, or the training setup).

> This model relies entirely on manually engineered MFCC features and does not directly learn representations from raw audio. That's fine for a baseline, and it lets us see how much information is already in the MFCC averages.

We train **two** classical models and keep the better one as our "baseline":

- **Logistic Regression** вЂ” a linear model, very fast, easy to interpret.
- **Random Forest** вЂ” an ensemble of decision trees, captures non-linear patterns, usually stronger than Logistic Regression on small tabular datasets.

We then evaluate both with:

- accuracy
- macro F1-score (better than accuracy when classes are slightly imbalanced)
- a full classification report (precision/recall/F1 per emotion)
- a confusion matrix (which emotions get confused with which)

````python
# --- Logistic Regression ---
log_reg = LogisticRegression(
    max_iter=2000,
    multi_class="multinomial",   # use proper multi-class loss
    solver="lbfgs",
    random_state=SEED,
)
log_reg.fit(X_train_scaled, y_train)

y_pred_lr = log_reg.predict(X_test_scaled)
acc_lr = accuracy_score(y_test, y_pred_lr)
f1_lr = f1_score(y_test, y_pred_lr, average="macro")
print(f"Logistic Regression вЂ” accuracy: {acc_lr:.4f}, macro F1: {f1_lr:.4f}")
````

````python
# --- Random Forest ---
rf = RandomForestClassifier(
    n_estimators=300,
    random_state=SEED,
    n_jobs=-1,         # use all CPU cores
)
rf.fit(X_train_scaled, y_train)

y_pred_rf = rf.predict(X_test_scaled)
acc_rf = accuracy_score(y_test, y_pred_rf)
f1_rf = f1_score(y_test, y_pred_rf, average="macro")
print(f"Random Forest        вЂ” accuracy: {acc_rf:.4f}, macro F1: {f1_rf:.4f}")
````

````python
# Pick whichever model did better вЂ” that's our official baseline.
if f1_rf >= f1_lr:
    baseline_name = "Random Forest"
    baseline_model = rf
    y_pred_baseline = y_pred_rf
    acc_baseline = acc_rf
    f1_baseline = f1_rf
else:
    baseline_name = "Logistic Regression"
    baseline_model = log_reg
    y_pred_baseline = y_pred_lr
    acc_baseline = acc_lr
    f1_baseline = f1_lr

print(f"Selected baseline: {baseline_name}")
print(f"  Accuracy: {acc_baseline:.4f}")
print(f"  Macro F1: {f1_baseline:.4f}")
````

````python
# Full classification report for the chosen baseline
print(f"Classification report вЂ” {baseline_name}\n")
print(classification_report(y_test, y_pred_baseline, target_names=le.classes_))
````

````python
# Confusion matrix for the chosen baseline
cm_baseline = confusion_matrix(y_test, y_pred_baseline)
plt.figure(figsize=(7, 5))
sns.heatmap(cm_baseline, annot=True, fmt="d", cmap="Blues",
            xticklabels=le.classes_, yticklabels=le.classes_)
plt.title(f"Confusion matrix вЂ” {baseline_name} (baseline)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()
````

**Interpreting the baseline.**

- *Strengths.* Tree-based models like Random Forest deal naturally with non-linear relationships between MFCC values and emotions, and they don't need much hyperparameter tuning to be reasonable. Logistic Regression is even simpler and trains in seconds.
- *Weaknesses.* Both models treat the 40 averaged MFCCs as the entire description of a clip вЂ” all temporal structure is collapsed away. Emotions that depend on *how the voice changes over time* (rising intonation in fear vs. surprise, for example) are harder to tell apart this way.
- *Overfitting?* Random Forest is more prone to overfitting on small datasets, especially with deep trees. If we see training accuracy near 100% but test accuracy much lower, that's the signal. We will keep an eye on it in the comparison section.

## 8. Model 2 вЂ” Dense Neural Network (MLP)

A **Multilayer Perceptron** is the simplest "deep learning" model: just a few fully-connected layers stacked on top of each other. It takes the same 40-dimensional MFCC mean vector as input, so this is a fair head-to-head comparison with the baseline.

We use:

- **Two hidden layers** with 256 and 128 units, ReLU activation, which lets the model learn non-linear combinations of MFCCs.
- **Dropout** between layers. Dropout randomly turns off some neurons during training, which prevents the network from becoming too reliant on any single feature combination and reduces overfitting.
- **Softmax output** over the 8 emotion classes.
- **Sparse categorical cross-entropy** loss, since our labels are integer-encoded.
- **Adam** optimizer, a reasonable default that doesn't need much tuning.
- **EarlyStopping** so we stop training as soon as the validation loss stops improving вЂ” that gives us a free anti-overfitting mechanism.

> Why might a neural network do better than the baseline? Because it can learn *interactions* between MFCC coefficients вЂ” for example "high mfcc_2 AND low mfcc_5 means angry" вЂ” that linear models can't represent and that even Random Forest may miss if the split isn't obvious. The flip side is that NNs need more data; with ~1,440 samples in RAVDESS, the improvement may be modest.

````python
num_classes = len(le.classes_)
input_dim = X_train_scaled.shape[1]

model = Sequential([
    Input(shape=(input_dim,)),
    Dense(256, activation="relu"),
    Dropout(0.3),
    Dense(128, activation="relu"),
    Dropout(0.3),
    Dense(num_classes, activation="softmax"),
])

model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"],
)

model.summary()
````

````python
# Early stopping watches the validation loss and stops when it stops improving.
# `restore_best_weights=True` rolls the model back to the best epoch, which is
# safer than just stopping at the last epoch.
early_stop = EarlyStopping(
    monitor="val_loss",
    patience=8,
    restore_best_weights=True,
    verbose=1,
)

history = model.fit(
    X_train_scaled, y_train,
    validation_split=0.2,   # carve out 20% of training as validation
    epochs=50,
    batch_size=32,
    callbacks=[early_stop],
    verbose=2,
)
````

## 9. Evaluate the deep learning model

We plot the four training curves (training/validation accuracy and loss) to **see** overfitting rather than just measure it. If the training curve keeps improving while the validation curve stalls, that's overfitting.

````python
hist = history.history

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(hist["accuracy"], label="train accuracy")
axes[0].plot(hist["val_accuracy"], label="val accuracy")
axes[0].set_title("Accuracy across epochs")
axes[0].set_xlabel("Epoch")
axes[0].set_ylabel("Accuracy")
axes[0].legend()

axes[1].plot(hist["loss"], label="train loss")
axes[1].plot(hist["val_loss"], label="val loss")
axes[1].set_title("Loss across epochs")
axes[1].set_xlabel("Epoch")
axes[1].set_ylabel("Loss")
axes[1].legend()

plt.tight_layout()
plt.show()
````

**How to read these curves.** We can see validation accuracy stop improving while training accuracy continues increasing, which suggests the model is beginning to overfit. EarlyStopping kicks in and rolls back to the best validation epoch, so the final model we evaluate is *not* the over-fit one.

````python
# Test-set evaluation
test_loss, test_acc = model.evaluate(X_test_scaled, y_test, verbose=0)
y_pred_nn_probs = model.predict(X_test_scaled, verbose=0)
y_pred_nn = np.argmax(y_pred_nn_probs, axis=1)

f1_nn = f1_score(y_test, y_pred_nn, average="macro")
print(f"Dense NN вЂ” test accuracy: {test_acc:.4f}, macro F1: {f1_nn:.4f}")
````

````python
print("Classification report вЂ” Dense Neural Network\n")
print(classification_report(y_test, y_pred_nn, target_names=le.classes_))
````

````python
cm_nn = confusion_matrix(y_test, y_pred_nn)
plt.figure(figsize=(7, 5))
sns.heatmap(cm_nn, annot=True, fmt="d", cmap="Greens",
            xticklabels=le.classes_, yticklabels=le.classes_)
plt.title("Confusion matrix вЂ” Dense Neural Network")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()
````

**Interpreting the deep learning model.**

- Looking at the confusion matrix, we can typically see emotions like *happy* vs. *surprised* and *fearful* vs. *sad* getting confused вЂ” these tend to have similar spectral profiles when temporal information is collapsed.
- The neural network often improves slightly over the baseline. With only ~1,440 samples and 40 averaged MFCCs as the only input, the ceiling for this architecture is fairly low. The CNN, LSTM, and CNN-LSTM models our teammates will build are expected to do significantly better because they can use the *temporal* structure of the MFCCs instead of throwing it away.

## 11. Assumptions and limitations

It's important to be honest about what this pipeline does and doesn't capture, so anyone reading our final report knows where the results stop being trustworthy.

- **MFCCs are a lossy summary.** They are designed around human hearing perception, which is a reasonable inductive bias for speech, but they're not the only useful representation. Raw spectrograms or learned representations (wav2vec, etc.) often outperform MFCCs.
- **We average MFCCs across time.** This is the biggest information loss in the pipeline. A clip where the voice rises in pitch over time looks identical, after averaging, to one where it falls. The temporal sequence models our teammates build will fix this.
- **The dataset is small.** RAVDESS has about 1,440 speech clips. Neural networks generally benefit from far more data than that, so we shouldn't expect a deep model to outperform Random Forest by a huge margin here.
- **Acted emotions are not real emotions.** RAVDESS actors deliberately portray emotions, often more clearly than people do in real conversations. Models trained on RAVDESS tend to do worse on "in the wild" speech.
- **Speaker leakage.** Our random split lets the same speaker appear in train and test, so part of what we are measuring is "can we recognize the speaker", not just "can we recognize the emotion". The actor-aware split in section 6 is the more honest evaluation.
- **Emotion confusion is built-in.** Some emotions are inherently easy to confuse (calm vs. neutral; fearful vs. surprised). No model trained on this data will be perfect on those pairs.
- **Dataset bias.** All 24 RAVDESS actors are English-speaking adults with North American accents. The model will not generalize cleanly to other languages or age groups.
- **Limited generalization.** Everything above means we should treat our reported metrics as upper bounds on what would happen in a real product вЂ” not as production-ready numbers.

## 12. Future work

The whole point of this notebook is to set up the comparison the team will do next. Concrete next steps:

- **Sequence models.** LSTM or GRU over the full `(40, n_frames)` MFCC matrix would keep the temporal information we are currently averaging away.
- **CNN on spectrograms.** Treat the spectrogram as an image and use convolutional layers. This tends to be one of the strongest approaches for audio classification.
- **CNN-LSTM hybrid.** Use a CNN to extract local spectral patterns and an LSTM on top to model how those patterns evolve in time. This is often the best of both worlds.
- **Transformer-based audio models.** Pretrained models like wav2vec 2.0 or HuBERT, fine-tuned on RAVDESS, would almost certainly outperform anything in this notebook.
- **Data augmentation.** Adding small amounts of noise, shifting pitch, or stretching the time axis effectively multiplies the dataset and tends to improve generalization a lot on small audio datasets.
- **Hyperparameter tuning.** Grid search or Optuna over learning rate, dropout, hidden layer sizes.
- **Cross-validation.** A single 80/20 split is noisy on a dataset this small. K-fold CV (especially actor-stratified K-fold) would give us much more reliable numbers.
- **Larger, more realistic datasets.** Combining RAVDESS with IEMOCAP, CREMA-D, or TESS вЂ” or fine-tuning on those instead вЂ” would give a model that works on more than just acted speech.

## 13. Report-style summaries

Short, copy-pasteable paragraphs for the final write-up.

### EDA summary

We started with about 1,440 `.wav` files spread across 24 actors and 8 emotion classes. The dataset is *almost* balanced, except that the *neutral* class has roughly half as many recordings as the others because RAVDESS doesn't include a "strong" intensity version of neutral. Clip durations are tightly clustered around 3вЂ“4 seconds, and all files use a consistent 48 kHz sampling rate. Waveform and spectrogram comparisons show clearly different acoustic patterns across emotions, which is encouraging: there is signal in the audio for a model to pick up on.

### Data cleaning summary

We checked that every file ended in `.wav`, that every filename followed the RAVDESS 7-part naming convention, and that every emotion code was one of the 8 valid codes. We also asked librosa to open every file as a sanity check, and dropped any that failed. After all of these checks, we re-ran one more validation step right after feature extraction to catch any `NaN` or `inf` values in the MFCC outputs.

### Preprocessing summary

We encoded the 8 emotion labels as integers, did a stratified 80/20 train/test split so each emotion is represented proportionally, and standardized the features with `StandardScaler`. Crucially, the scaler was fit on the training set only and then applied to the test set, to avoid leakage. We also implemented (but did not use by default) an actor-aware split that holds out entire actors for testing вЂ” that is the more honest measurement of how well the model generalizes to new speakers.

### Baseline model summary

We compared Logistic Regression and Random Forest as classical baselines, and kept the stronger of the two. With 40 time-averaged MFCC features, the baseline typically reaches a respectable accuracy on RAVDESS. Its main weakness is that it only sees a single 40-dim vector per clip, so any information that lives in *how* the voice changes over time is invisible to it.

### Dense neural network summary

We trained a 2-hidden-layer MLP (256 в†’ 128, with dropout 0.3 between layers) for up to 50 epochs with Adam, sparse categorical cross-entropy loss, and EarlyStopping watching the validation loss. EarlyStopping restored the best validation weights, which protects us from the over-fit final epochs.

### Overfitting discussion

The training accuracy curve continues climbing past the point where the validation accuracy plateaus вЂ” the classic shape of a model that has started memorizing the training set. Dropout slows this down but doesn't eliminate it on a dataset this small. EarlyStopping is what actually saves us: it rolls the weights back to the best validation epoch, so the model we evaluate is *not* the overfit one.

### Final results

Both models do meaningfully better than chance (12.5% for 8 balanced classes). The Dense NN tends to edge out the baseline by a few points, but with 40 averaged MFCCs as the only input, neither model can reach the high-90s accuracy that papers using temporal models report on RAVDESS. That gap is exactly what the CNN, LSTM, and CNN-LSTM teammates will close.

### Limitations

The dataset is small, the speakers are all English-speaking adults, the emotions are acted (and so cleaner than real-world speech), and our default split allows the same speaker into both train and test. We average MFCCs across time, which throws away the temporal dynamics that distinguish many emotions (calm vs. sad, fearful vs. surprised). Everything we report should be read as an upper bound.

### Future work

Move to sequence models (LSTM/GRU), spectrogram-CNNs, CNN-LSTM hybrids, and ideally a pretrained transformer (wav2vec / HuBERT) fine-tuned on RAVDESS. Add data augmentation, do K-fold cross-validation with actor-stratified folds, and consider combining RAVDESS with other emotion datasets so the final system generalizes beyond acted North American English speech.
