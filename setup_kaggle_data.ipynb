## Download RAVDESS Dataset

```python
import os

DATA_DIR = '/content/data'
os.makedirs(DATA_DIR, exist_ok=True)

!kaggle datasets download -d uwrfkaggler/ravdess-emotional-speech-audio -p {DATA_DIR} --unzip

print(f"\nDataset downloaded to: {DATA_DIR}")
print(f"Contents:")
!ls -la {DATA_DIR}
```

## Verify Download

```python
import glob

audio_files = glob.glob(f'{DATA_DIR}/**/*.wav', recursive=True)
print(f"Total audio files found: {len(audio_files)}")

if audio_files:
    print(f"\nSample files:")
    for f in audio_files[:5]:
        print(f"  {f}")
```
